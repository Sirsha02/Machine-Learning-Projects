```python
#Downloading the datafile
import subprocess
import zipfile
import os

# Competition name
competition = "dogs-vs-cats"

# Download folder
download_dir = "."


print(f"Downloading competition data for: {competition}")

# Use utf-8 to avoid UnicodeDecodeError
process = subprocess.Popen(
    ["kaggle", "competitions", "download", "-c", competition, "-p", download_dir],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    encoding="utf-8"  # instead of text=True
)

# Print output line by line
for line in process.stdout:
    print(line.strip())

process.wait()

if process.returncode == 0:
    print("✅ Download finished!")

    # Find ZIP file
    for file in os.listdir(download_dir):
        if file.endswith(".zip"):
            zip_path = os.path.join(download_dir, file)
            extract_dir = os.path.join(download_dir, "unzipped")

            print(f"Unzipping {file} to {extract_dir} ...")
            with zipfile.ZipFile(zip_path, 'r') as zip_ref:
                zip_ref.extractall(extract_dir)

            print(f"✅ Unzipped to: {extract_dir}")

else:
    print("❌ Download failed. Check your Kaggle setup.")
```

```python
#REVISED CODE FOR THE ABOVE, SINCE FIRST 2K PHOTOS ARE ONLY CAT

import os
from PIL import Image

original_folder = 'train/'
resized_folder = 'image resized/'

# Make sure output folder exists
os.makedirs(resized_folder, exist_ok=True)

# List all images
all_files = os.listdir(original_folder)

# Separate cats and dogs
cat_files = [f for f in all_files if f.startswith('cat')]
dog_files = [f for f in all_files if f.startswith('dog')]

# Pick 1000 cats and 1000 dogs
cat_files = cat_files[:1000]
dog_files = dog_files[:1000]

# Combine them
selected_files = cat_files + dog_files

print(f"Resizing {len(selected_files)} images: {len(cat_files)} cats and {len(dog_files)} dogs.")

for filename in selected_files:
    img_path = os.path.join(original_folder, filename)

    img = Image.open(img_path)
    img = img.resize((224, 224))
    img = img.convert('RGB')

    new_img_path = os.path.join(resized_folder, filename)
    img.save(new_img_path)

print("✅ Done resizing!")

```

```python
#creating labels for resized images of dogs and cats
#cat -> 0 , dog -> 1

filenames = os.listdir('image resized')

labels = []

for i in range (2000):
    file_name = filenames[i]
    label = file_name[0:3]

    if label == 'dog':
        labels.append(1)

    else:
        labels.append(0)  

```

```python
#converting resized images to numpy arrays

import cv2
import glob

image_directory = 'image resized/'
image_extension = ['png','jpg']

files = []

[files.extend(glob.glob(image_directory + '*.' + e)) for e in image_extension]

dog_cat_images = np.asarray([cv2.imread(file) for file in files])
```

```python
X = dog_cat_images
Y = np.asarray(labels)
```

```python
#train test split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

```python
#scaling the data to 0-1

x_train_scaled = x_train/255

x_test_scaled = x_test/255
```

```python
print(x_train_scaled)
print(x_test_scaled)
```

```python
import tensorflow as tf
import tensorflow_hub as hub
import tf_keras
```

```python
mobilenet_model = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'

model = tf_keras.Sequential([
    hub.KerasLayer(mobilenet_model, input_shape=(224,224,3), trainable=False)
])

num_of_classes = 2

model.add(tf_keras.layers.Dense(num_of_classes, activation='softmax'))

model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics = ['acc']
)
```

```python
model.fit(x_train_scaled, y_train, epochs=5)
```

```python
score, acc = model.evaluate(x_test_scaled, y_test)
print('Test Loss =', score)
print('Test Accuracy =', acc)
```

```python
import cv2
import numpy as np

# Load image path
input_image_path = input('Path of the image to be predicted: ')

# Read image
input_image = cv2.imread(input_image_path)

# Display image (cv2.imshow works locally)
cv2.imshow('Input Image', input_image)
cv2.waitKey(0)  # Waits for any key press
cv2.destroyAllWindows()  # Closes the window

# Resize and preprocess
input_image_resize = cv2.resize(input_image, (224, 224))
input_image_scaled = input_image_resize / 255.0
image_reshaped = np.reshape(input_image_scaled, [1, 224, 224, 3])

# Make prediction
input_prediction = model.predict(image_reshaped)
print(input_prediction)

# Get predicted label
input_pred_label = np.argmax(input_prediction)
print(input_pred_label)

# Interpret result
if input_pred_label == 0:
    print('The image represents a Cat')
else:
    print('The image represents a Dog')
```
